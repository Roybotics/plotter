```python
def morton(x, y):
    x = (x | (x << 16)) & 0x0000FFFF0000FFFF;
    x = (x | (x << 8))  & 0x00FF00FF00FF00FF;
    x = (x | (x << 4))  & 0x0F0F0F0F0F0F0F0F;
    x = (x | (x << 2))  & 0x3333333333333333;
    x = (x | (x << 1))  & 0x5555555555555555;

    y = (y | (y << 16)) & 0x0000FFFF0000FFFF;
    y = (y | (y << 8))  & 0x00FF00FF00FF00FF;
    y = (y | (y << 4))  & 0x0F0F0F0F0F0F0F0F;
    y = (y | (y << 2))  & 0x3333333333333333;
    y = (y | (y << 1))  & 0x5555555555555555;

    return x | (y << 1);


#ag.genCircles.sort(key=lambda x: morton(int(x.m.real*16),int(x.m.imag*16)), reverse=False)
```

```python
r1 = 4
r2 = 4
r3 = 4

ag = ApollonianGasket(1/r1, 1/r2, 1/r3)
ag.generate(30,minRadius=0.02)
#print(ag.genCircles)


#print((ag.genCircles))
print(len(ag.genCircles))
print("done!")
```

```python
plotter = XYPlotter()
plotter.connect()
```

```python
plotter.pendown()
```

```python
ag.genCircles.sort(key=lambda x: x.r.real,        reverse=True)
#ag.genCircles.sort(key=lambda x: int(x.m.real*4), reverse=False)
#ag.genCircles.sort(key=lambda x: int(x.m.imag*4), reverse=False)
#ag.genCircles.sort(key=lambda x: morton(int(x.m.real*16),int(x.m.imag*16)), reverse=False)
ag.genCircles = [circle for circle in ag.genCircles if circle.r.real > 0.02]

x0 = 11
y0 = 13
plotter.gotozero()
for circle in ag.genCircles:
    print (circle)
    x = x0+circle.m.real
    y = y0+circle.m.imag
    r = circle.r.real
    speed = 2000 if r<0.2 else ( 1000 if r<0.5 else ( 500 if r<1 else 300)) 
    plotter.circle(x,y,r,dt=speed)
    if r>0.1:
        plotter.circle(x,y,r,dt=speed)
    radius = 0.381*r
    revolutions = 14*radius
    if revolutions>0 and r>0.1:
        plotter.spiral(x,y,0,radius,0,0,revs=revolutions,dt=speed)

plotter.gotozero()
plotter.penOff()
```
